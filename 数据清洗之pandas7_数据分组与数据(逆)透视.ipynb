## groupby与agg函数的连用  
**传入一列数， 返回一个标量, 接受的函数的返回值必须是一个标量值或者聚合值**

```python
gp = df.groupby(by='部门名称')
gp
```

在 DataFrame 的分组上，传递一个应用于每个列的函数列表：

### 同时对多个列运行多个函数

```python
gp.agg([np.mean, np.sum, np.max])  #对dataframe中所有数值型的字段进行聚合运算
```

### 单独对某个列运行多个函数

当然也可以给分组后的某个字段单独的看多个聚合函数的情况

```python
gp['金额1'].agg([np.mean, np.sum, np.max])
```

### 对不同的列执行不同的聚合函数

```python
A = gp.agg({'金额1':[np.mean, np.max, np.min], 
        '金额2':[np.ptp],
        '金额3':[np.std, np.var],
        '金额4':['sum', 'mean']    # 等价于[np.sum, np.mean]
       })
A
```

```python
a = gp.agg({'金额1':[np.mean, np.max, np.min], 
        '金额2':[np.ptp],
        '金额3':[np.std, np.var],
        '金额4':['sum', 'mean']   
       })
```

```python
a
```

```python
a.reset_index(level=0, col_fill="哈哈").droplevel(level=0,axis=1)
```

### agg接受自定义函数    
**要求：该函数的返回值必须是一个标量**  
自定义一个求最大值和最小值和的函数

```python
def X1(s):
    s_ = list(set(s))
    s_.sort()
    return s_[0]+s_[-1]
```

```python
gp.get_group('人力部').金额1
```

```python
X1(gp.get_group('人力部').金额1)
```

```python
gp.agg({'金额1':X1})
```

如果返回的是一个序列， 一个Series就会报错

```python
def X2(s):
    s_ = list(set(s))
    s_.sort()
    return pd.Series([s_[0], s_[-1]])                             # 返回的是一个DataFrame
```

```python
X2(gp.get_group('人力部').金额1)
```

```python
gp.agg({'金额1':X2})  # 这里虽然没有报错， 但是现实把结果破坏了
```

**报错说必须产生一个聚合值**

## groupby与apply函数的连用  
- 该函数的返回值可以标量， 也可以是一个Series, 或者一个DataFrame
- apply是对每个组进行操作， 每个组就是一个DataFrame

```python
a = gp['金额1'].apply(X2)
a
```

```python
gp.apply(lambda x:x.describe())
```

## 对所有列执行相同的聚合

```python
a = df.pivot_table(index='月', columns='部门名称', aggfunc="sum")
a
```

## 指定对某些列进行何种聚合

```python
a = df.pivot_table(index='月', columns='部门名称', aggfunc={'金额1':np.sum, 
                                                           '金额2':[np.mean, 'max']                                                              
                                                                })
a
```
