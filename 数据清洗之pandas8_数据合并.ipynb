```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(666)
a = np.random.randint(60, 100, (5, 3))
df1 = pd.DataFrame(a, 
                   index=['A1', 'A2', 'A3', 'A4', 'A5'], 
                   columns=['语文', '数学', '英语'])
b = np.random.randint(60, 100, (3, 4))
df2 = pd.DataFrame(b, 
                   index=['A6', 'A7', 'A8'], 
                   columns=['物理', '数学', '英语', '语文'])
c = np.random.randint(60, 100, (3, 2))
df3 = pd.DataFrame(c, 
                   index=['A2', 'A3', 'A9'], 
                   columns=['物理', '化学'])

df1
df2
df3
```

# pd.concat([df1, df2，df3..], axis=0, join='outer', sort=False)  
**axis=0 默认是纵向拼接 , axis=1横向拼接。**   
**join默认连接方式是外连接, 当然还有inner内连接。注意这里没有左连接和右连接。**  
**sort=False只是消除警告而已**

## 纵向外连接  
axis=0代表纵向， 会按照相同的列名进行拼接。  
join='outer'表示外连接， 外连接会保留所有的信息， 匹配不上的地方就是缺失值。

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=0,join='outer', sort=False)
```

## 纵向内连接  
axis=0  
内连接是保留匹配上的数据， 纵向连接的时候， 是以列来匹配的， 即保留大家都有的列

```python
df1
df2
df3
```

```python
pd.concat([df1, df3], axis=0, join='inner', sort=False)  # 没有相同的列名
```

```python
pd.concat([df1, df2], axis=0,join='inner',  sort=False)
```

## 横向外连接  
axis=1  
外连接保留两张表所有的信息

```python
df1
df2
df3
```

```python
pd.concat([df1, df2], axis=1, join='outer', sort=False)  # 保留所有的信息
```

## 横向内连接  
axis=1
join='inner', 好留两张表都有的行索引

```python
df1
df2
df3
```

```python
pd.concat([df1, df2], axis=1, join='inner', sort=False)
```

```python
pd.concat([df1, df3], axis=1, join='inner', sort=False)
```

# pd.merge(左表, 右表，how=连接方式， on=连接字段 )  
**pd.merge()基本参照的是数据库表的连接方式， 也是用的最多的连接方式， 逻辑上与数据库保持一致。**  
**how='inner', 即默认是内连接， 这里还可以选择外连接outter 左连接left  右连接right**

```python
df1 = df1.reset_index().rename(columns={'index':'姓名'})
df1
```

```python
df2 = df2.reset_index().rename(columns = {'index':'姓名'})
df2
```

```python
df3 = df3.reset_index().rename(columns = {'index':'姓名'})
df3
```

```python
df1
df2
df3
```

## 内连接  
how='inner'  
保留在连接字段上大家都有的

```python
pd.merge(df1, df2, how='inner', on='姓名')
```

```python
pd.merge(df1, df3, 'inner', '姓名')
```

## 外连接  
how='outer'  
全部所有的信息

```python
df1
df2
df3
```

```python
pd.merge(df1, df2, how='outer', on='姓名')
```

```python
pd.merge(df1, df3, how='outer', on='姓名')
```

## 左连接  
how='left'  
保留左表所有的信息

```python
df1
```

```python
df3
```

```python
pd.merge(df3, df1, how='left', on='姓名')
```

## 右连接  
how = 'right'  
保留右表所有的信息

```python
pd.merge(df1, df3, 'right', '姓名')
```
