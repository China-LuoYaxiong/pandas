```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randint(0, 6, (10, 6)))
```

```python
# 用每行的前6个数据，预测后面的数据
L = []
for i in ["ym"+str(i) for i in range(1,7)]:
    # 计算第一次的均值
    x = df.mean(axis=1)
    # 保存到列表里
    L.append(x)
    # df.shift(-1, axis=1)把数据整体先向左边移动一位，导致第一列删除，最后一列为空。
    # .dropna(axis=1)是把为空的列删除，其实就是把最后一列删除，
    df = df.shift(-1, axis=1).dropna(axis=1)
    # 把新的数据添加到最后一列，就和原来的数据列数一样多， 相当于把新得到的结果添加到最后， 同时删除最前的那一列
    df[i] = x
    # 开始循环的做预测
```
